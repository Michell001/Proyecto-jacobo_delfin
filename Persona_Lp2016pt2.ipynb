```python
import pandas as pd #para manejar el dataframe
import numpy as np #para abrir el csv
import matplotlib.pyplot as plt #para graficar 
from sklearn.preprocessing import normalize #para normalizar los datos
from sklearn.cluster import AgglomerativeClustering
```

```python
path = 'C:/Users/sala5/Desktop/Persona-Lp2016ff.csv'
df = pd.read_csv(path, encoding='latin', low_memory=False)
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
df
```

```python
df_sample = df.sample(frac=0.05, random_state=1)
df_sample.head()
```

```python
x = normalize(df_sample, norm='l2', axis=1, copy=True)#,return_norm=False)

x.shape
```

```python
ward = AgglomerativeClustering(n_clusters=4, 
                               affinity='euclidean', 
                               linkage='ward')
```

```python

y_w = ward.fit_predict(x)
```
